# Pattern Definition

Before we can match any queues generated by scanning to SP or HP patterns, we must store the information about the patterns themselves.

```python
from collections import namedtuple
```

```python
SP1 = ['SP1','efftte']
SPs = [
    ['SP1','efftte'],
    ['SP1r','ettffe'],
    ['SP2','eftfte'],
    ['SP2r','etftfe'],
    ['SP3','efttfe'],
    ['SP4','etffte']]
```

We must be able to extract the necessary matching pattern for each color and to extract the relative locations of triggers.

```python
EMPTY_CHAR = '+'
PLAYER_CHARS = {'red':'r',
                     'black':'b'}

PATTERN_TEMPLATE = namedtuple('Pattern','name match_pat triggers')

def pattern_processor(pattern, player_marker):
    processed_patterns = []
    for name, pattern in SPs:
        matching_pattern = []
        trigger_locs = []
        for index, char in enumerate(pattern):
            if char == 'e':
                matching_pattern.append(EMPTY_CHAR)
            elif char == 't':
                matching_pattern.append(EMPTY_CHAR)
                trigger_locs.append(index-5)
            elif char == 'f':
                matching_pattern.append(player_marker)
        processed_patterns.append(PATTERN_TEMPLATE(name,matching_pattern,trigger_locs))
    return processed_patterns
```

```python
red_SP_patterns = pattern_processor(SPs, 'r')
display(red_SP_patterns)
```
