# Matrix Conversion

Let's convert the game's df to a matrix mimicking the game board itself by:

1. Generating an empty matrix the size of the board

```python
import numpy as np
```

```python
empty_char = '+'
height = int(game.right_edge-game.left_edge)
width = int(game.top_edge-game.bottom_edge)
matrix = np.ones((height,width),dtype=str)
matrix[:,:] = empty_char
display(matrix)
```

2. Transforming the coordinates of the df to match the matrix

```python
df = game.df.copy(deep=True)
df['i'] = (game.top_edge-0.5)-df['y']
df['j'] = df['x'] - (game.left_edge+0.5)
display(df)
```

3. Writing the red and black moves into the empty matrix.

```python
conv_dict = {'black':'b',
             'red':'r'}
df_mat = df[['marker','i','j','player']]
for rowi in range(df_mat.shape[0]):
    marker, i, j, player = df_mat.iloc[rowi,:]
    matrix[int(i),int(j)] = conv_dict[player]
display(matrix)
game.render()
```

# Matrix Scanning

We will first try to scan the columns of the matrix.

```python
height, width = matrix.shape
for j in range(width):
    queue = ['_']*5
    for i in range(height):
        queue.pop(0)
        queue.append(matrix[i,j])
        print(queue)
        if queue== ['r']*5:
            print('red won!')
    
```

We will scan the rows as well

```python
height, width = matrix.shape
for i in range(height):
    queue = ['_']*5
    for j in range(width):
        queue.pop(0)
        queue.append(matrix[i,j])
        print(queue)
        if queue== ['b']*5:
            print('black won!')
```

Now we must scan diagonals, which has a few steps:

1. Create a list of all points on the left and top edges of the matrix.

```python
I,J = matrix.shape # Height and width of matrix
left_points = zip(list(range(I-1,0,-1)),[0]*I)
top_points = zip([0]*J,list(range(0,J+1)))
start_points = list(left_points) #copy the list
start_points.extend(top_points)
display(start_points)
```

2. Loop through all those edge points and move down and to the right the correct amount

```python
for i_0,j_0 in start_points:
    queue = ['_']*5
    N = min(I-i_0, J-j_0)
    print(f'N = {N}')
    for n in range(N):
        queue.pop(0)
        queue.append(matrix[i_0+n, j_0+n])
        print(queue)
        if queue == ['r']*5:
            print('red won diagonally!')
```
